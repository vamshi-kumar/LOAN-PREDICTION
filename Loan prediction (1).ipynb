```python
df['LoanAmount_log'] = np.log(df['LoanAmount'])
df['LoanAmount_log'].hist(bins=20)
```

```python
df['Gender'] = df['Gender'].map({'Female':0,'Male':1}).astype(np.int)
df['Married'] = df['Married'].map({'No':0, 'Yes':1}).astype(np.int)
df['Education'] = df['Education'].map({'Not Graduate':0, 'Graduate':1}).astype(np.int)
df['Self_Employed'] = df['Self_Employed'].map({'No':0, 'Yes':1}).astype(np.int)
df['Loan_Status'] = df['Loan_Status'].map({'N':0, 'Y':1}).astype(np.int)
df['Dependents'] = df['Dependents'].astype(np.int)
df['Property_Area'] = df['Property_Area'].map({'Urban':0, 'Semiurban':1, 'Rural':2}).astype(np.int)
```

```python
#for coverting into categorical variables into numeric
from sklearn.preprocessing import LabelEncoder
var_mod = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
```

```python
#Import models from scikit learn module:
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold   #For K-fold cross validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
```

```python
def classification_model(model, data, predictors, outcome):
    model.fit(data[predictors],data[outcome])
    predictions = model.predict(data[predictors])
    accuracy = metrics.accuracy_score(predictions,data[outcome])
    print ("Accuracy : %s" % "{0:.3%}".format(accuracy))
    kf = KFold(data.shape[0], n_folds=5)
    error = []
    for train, test in kf:
        train_predictors = (data[predictors].iloc[train,:]) # Filter training data
        train_target = data[outcome].iloc[train]     # Target data used to train algorithm
        model.fit(train_predictors, train_target)
        error.append(model.score(data[predictors].iloc[test,:], data[outcome].iloc[test]))
    print ("Cross-Validation Score : %s" % "{0:.3%}".format(np.mean(error)))
    model.fit(data[predictors],data[outcome])
```

```python
#logistic Regression
outcome_var = 'Loan_Status'
model1 = LogisticRegression()
```

```python
predictor_var1 = ['Education']
classification_model(model1, df,predictor_var1,outcome_var)
```

```python
predictor_var2 = ['Married']
classification_model(model1, df,predictor_var2,outcome_var)
```

```python
predictor_var3 = ['Self_Employed']
classification_model(model1, df,predictor_var3,outcome_var)
```

```python
predictor_var4 = ['Property_Area']
classification_model(model1, df,predictor_var4,outcome_var)
```

```python
predictor_var5 = ['Credit_History']
classification_model(model1, df,predictor_var5,outcome_var)
```

```python
predictor_var6 = ['Credit_History','Education','Married','Self_Employed','Property_Area']
classification_model(model1, df,predictor_var6,outcome_var)
```

```python
#Decision Tree
model2 = DecisionTreeClassifier()
predictor_var7 = ['Credit_History','Gender','Married','Education','Self_Employed','Dependents']
classification_model(model2, df,predictor_var7,outcome_var)
```

```python
predictor_var8= ['Credit_History','Loan_Amount_Term','LoanAmount_log']
classification_model(model2, df,predictor_var8,outcome_var)
```

```python
#Random Forest
model3 = RandomForestClassifier(n_estimators=100)
predictor_var9 = ['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'Loan_Amount_Term', 'Credit_History', 'Property_Area',
        'LoanAmount_log','TotalIncome_log']
classification_model(model3, df,predictor_var9,outcome_var)
```
